# Fluid Motion

## The Blasius Boundary Layer

First, congratulations on making it to the 5th module of #numericalmooc! It's been quite a journey. In this notebook, we are going to be applying our knowledge to the complex world of fluid motion. We are going to start with one of the simplest problems, which can be solved numerically, in fluids. As the title suggests, it's called the Blasius Boundary Layer.

In case some of you don't have a background in fluids, you can check out the [wikipedia page](http://en.wikipedia.org/wiki/Boundary_layer) on boundary layers to get an idea. Boundary layers can get very complex depending on the flow conditions and geometry of the surface, but today we are going to look at a very simple one.

Figure 1 shows fluid moving over a flat plate which continues infinitely in the positive x direction. The fluid moves with a unidirectional velocity we're calling $u_{\infty}$. This is the problem that Blasius solved as a student, a student of the founder of the field, but a student no less. 

![Image](./figures/boundary-layer.svg)
####Figure 1. Boundary layer over a flat plate.

The full derivation can be found [here](http://web.iitd.ac.in/~prabal/MEL242/%2814-15%29-Boundary%20layer%20theory.pdf), and the whole solution boils down to a single non-linear 3rd order Ordinary Differential Equation (ODE).

\begin{equation}
2 \frac{d^3 f}{d \eta^3} + f \frac{d^2 f}{d \eta^2} = 0
\end{equation}

Not bad right?

Now as you probably noticed, each term is differentiated with respect to $\eta$. And I'd imagine you're wondering what $\eta$ is. This term $\eta$ is the genius that made this solution possible. It's called a similarity variable, and the reason for the name will be explained in a little, but first we should get our assuption about this flow field on the table. 

Since we are working with an infinite flow field, we should talk in terms of non-dimensional values. In this case, Blasius defines a non-dimensional velocity term, $u/u_{\infty}$, where $u$ is the local velocity and $u_{\infty}$ is the velocity very far from the wall. He then defines a non-dimentional length term, $y/\delta$, where $y$ is the distance from the wall and $\delta$ is the [boundary layer thickness](http://en.wikipedia.org/wiki/Boundary_layer) as seen in Figure 1. 

The assuption that makes all of this work is that the *non-dimensional velocity will be constant for a given non-dimensional length.* That is, $u/u_{\infty}$ is constant for a given $y/\delta$.

That's it. That's the central assuption.

In the derivation that assuption comes through in the mathematics as the similarity variable $\eta$. 

\begin{equation}
\eta = \frac{y}{\sqrt{\nu x / u_{\infty}}}
\end{equation}

Okay, now that we have that on the table, we should cover some of the numerics. 

## The Shooting Method

The shooting method provides us with the simplest way to solve the Blasius Boundary Layer. So to get the idea suppose we have a Boundary Value Problem (BVP) with Dirichlet boundary conditions:

\begin{equation}
y'' = f(x,y,y'), \; y(a) = \alpha, \; y(b) = \beta .
\end{equation}

We then rearrange the BVP to form:

\begin{equation}
y' = z \\
z' = f(x,y,z) \\
y(a) = \alpha \\
y(b) = \beta.
\end{equation}

We can then turn the BVP above into an Initial Value Problem (IVP) by replacing the boundary condition $x = b$ with

\begin{equation}
z(a) = \theta,
\end{equation}

where $\theta$ is some number. We can then solve the IVP with the method of our choosing to obtain the solution at y(b). If $y(b) = \beta$ we've done it! We've solved the BVP! Mostly likely, though, it $y(b) \neq \beta$ on the first try and we will have to try other values of $\theta$ to before we find the right one. 

So what would it take to apply the shooting method to the problem we are working on?


### Boundary Conditions

Using the coordinate system shown in Figure 1, 

\begin{equation}
\textrm{At } x = 0: \; u(0,y) = u_{\infty},
\end{equation}

\begin{equation}
\textrm{At } y = 0: \; u(x,0) = u(0,y) = 0,
\end{equation}

\begin{equation}
\textrm{As } y \rightarrow \infty: \; u(x,\infty) = u_{\infty}.
\end{equation}

Pretty straight forward right?

Now, because this problem requires we work with a similarity variable ($\eta$) we should put these boundary conditions in terms of $\eta$.

\begin{equation}
\textrm{At } \eta = 0: \; f = 0,
\end{equation}

\begin{equation}
\textrm{At } \eta = 0: \; \frac{df}{d \eta} = 0,
\end{equation}

\begin{equation}
\textrm{As } \eta \rightarrow \infty: \; \frac{df}{d \eta} \rightarrow 1.
\end{equation}


### Converting to an IVP

As said the first step in implimenting the shooting method is turning the BVP into an IVP. To reiterate, we want to rewrite Equation 1 in such a way that it only involves first derivative and can therefore be solved as an IVP. In this case the three equations are

\begin{equation}
\frac{df}{d \eta} = g \; \textrm{with } f(0) = 0
\end{equation}

\begin{equation}
\frac{dg}{d \eta} = h \; \textrm{with } g(0) = 0, g(\infty)=1
\end{equation}

\begin{equation}
\frac{dh}{d \eta} = \frac{1}{2} f*h \; \textrm{with } h(0) = ?
\end{equation}

Now the problematic value is $h(0)$, but since we know $g(\infty) = 1$ we can use the shooting method to "shoot" for the third missing value.

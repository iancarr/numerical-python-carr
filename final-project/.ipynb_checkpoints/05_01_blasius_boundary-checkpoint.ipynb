# Fluid Motion

### The Blasius Boundary Layer

First, congradulations on making it to the 5th module of #numericalmooc! It's been quite a journey. In this notebook, we are going to be applying our knowledge to the complex world of fluid motion. We are going to start with one of the simplest problems, which can be solved numerically, in fluids. As the title suggests, it's called the Blasius Boundary Layer.

In case some of you don't have a background in fluids, you can check out the [wikipedia page](http://en.wikipedia.org/wiki/Boundary_layer) on boundary layers to get an idea. Boundary layers can get very complex depending on the flow conditions and geometry of the surface, but today we are going to look at a very simple one.

Imagine an infinite flat plate with an infinite field of fluid above it. The fluid moves with a unidirectional velocity we're calling $u_{\infty}$. This is the problem that Blasius solved as a student, a student of the founder of the field, but a student no less. 

The full derivation can be found [here](http://web.iitd.ac.in/~prabal/MEL242/%2814-15%29-Boundary%20layer%20theory.pdf), and the whole solution boils down to one 3rd order Ordinary Differential Equation (ODE).

\begin{equation}
2 \frac{d^3 f}{d \eta^3} + f \frac{d^2 f}{d \eta^2} = 0
\end{equation}

Not bad right?

Now as you probably noticed, each term is differentiated with respect to $\eta$. And I'd imagine you're wondering what $\eta$ is. This term $\eta$ is the genius that made this solution possible. It's called a similarity variable, and the reason for the name will be explained in a little, but first we should get our assuption about this flow field on the table. 

Since we are working with an infinite flow field, we should talk in terms of non-dimensional values. In this case, Blasius defines a non-dimensional velocity term, $u/u_{\infty}$, where $u$ is the local velocity and $u_{\infty}$ is the velocity very far from the wall. He then defines a non-dimentional length term, $y/\delta$, where $y$ is the distance from the wall and $\delta$ is the [boundary layer thickness](http://en.wikipedia.org/wiki/Boundary_layer). 

The assuption that makes all of this work is that the *non-dimensional velocity will be constant for a given non-dimensional length.* That is, $u/u_{\infty}$ is constant for a given $y/\delta$.

That's it. That's the central assuption.

In the derivation that assuption comes through in the mathematics as the similarity variable $\eta$. 

\begin{equation}
\eta = \frac{y}{\sqrt{\nu x / u_{\infty}}}
\end{equation}
